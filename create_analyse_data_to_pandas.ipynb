```python
import yfinance as yf
import pandas as pd

ticker = 'PETR4.SA'
data = yf.download(ticker,start="2024-01-01", end= "2024-12-23")
data.head()

data.to_csv('petr4.csv')

print("dados Petrobras 2024 baixados com sucesso :")
```

```python
print(pd.read_csv('petr4.csv'))
```

```python
df = pd.read_csv('petr4.csv')
df_cleaned = df[["Price","Open","High","Low","Close","Volume"]]
df_cleaned.dropna(inplace=True)
df_cleaned.to_csv('petr4_dados_limpos.csv', index=False)
df_cleaned.head(5)
df.head(5)
```

```python
df_cleaned.tail(5)
```

```python
print(df_cleaned["Volume"].dtype)
```

```python
# converter objeto para numerico
df_cleaned["Volume"] = pd.to_numeric(df_cleaned["Volume"], errors="coerce")
```

```python
print(df_cleaned["Volume"].dtype)
print(df_cleaned["Volume"].isnull().sum())
```

```python
df_cleaned.dropna(subset=["Volume"], inplace=True)
```

```python
top_10_volumes = df_cleaned.nlargest(10, "Volume")
top_10_volumes.head(10)
#print(top_10_volumes)
```
